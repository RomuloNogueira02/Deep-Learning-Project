```python
import os
```

# VAL

```python
for file in os.listdir("images/val"):
    filename = os.fsdecode(file)
    parts = file.rsplit('.png', 1)
    new_path = parts[0] + ".txt"
    before = os.getcwd().replace('\\','/') + '/labels/'+ new_path
    after = os.getcwd().replace('\\','/') + '/labels/val/'+new_path

    #print(before)
    #print(after)
    if os.path.exists(before):
        os.rename(before,after)
```
