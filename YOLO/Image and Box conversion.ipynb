## Convert images from 512 ->224

```python
current_path = os.getcwd()
```

```python
print(current_path)
```

```python
def resize_image(input_path, output_path, target_size=(224, 224)):
    with Image.open(input_path) as img:
        # Resize the image while maintaining the aspect ratio
        img.thumbnail(target_size, Image.ANTIALIAS)
        
        # Create a new image with a white background
        new_img = Image.new("RGB", target_size, (255, 255, 255))
        
        # Calculate the position to paste the resized image
        paste_position = (
            (target_size[0] - img.size[0]) // 2,
            (target_size[1] - img.size[1]) // 2
        )
        
        # Paste the resized image onto the white background
        new_img.paste(img, paste_position)
        
        # Save the resulting image
        new_img.save(output_path)
```

```python
image_path = current_path + '\images'
for file in os.listdir(image_path):
    filename = os.fsdecode(file)
    if len(filename) > 22:
        resize_image(image_path + '/' + filename, image_path + '/' + filename)
```

## Convert bounding boxes of the 512px images to the correct size

```python
import xml.etree.ElementTree as ET
```

```python
labels_path = current_path + '/labels/'
```

```python
def convert_to_yolo(xmin, xmax, ymin, ymax, img_width, img_height):
    x_center = (xmin + xmax) / (2 * img_width)
    y_center = (ymin + ymax) / (2 * img_height)
    
    box_width = (xmax - xmin) / img_width
    box_height = (ymax - ymin) / img_height
    
    return x_center, y_center, box_width, box_height
```

```python
def extract_bbox(xml_file,output_file):
    with open(output_file, 'w') as f:

        factor = 512 / 224
        tree = ET.parse(xml_file)
        root = tree.getroot()
        
        for obj in root.findall('.//object'):
            name = obj.find('name').text
            bbox = obj.find('bndbox')
            xmin = float(bbox.find('xmin').text)
            xmax = float(bbox.find('xmax').text)
            ymin = float(bbox.find('ymin').text)
            ymax = float(bbox.find('ymax').text)
            if len(xml_file)> 28:
                print("ORIGINAL: " + str(xmin) + " AFTER: " + str(xmin / factor))
                x_center, y_center, box_width, box_height = convert_to_yolo(xmin / factor, xmax / factor, ymin / factor, ymax / factor, 224, 224)
            else:
                x_center, y_center, box_width, box_height = convert_to_yolo(xmin, xmax, ymin, ymax, 224, 224)
            line = f"{0} {x_center} {y_center} {box_width} {box_height}\n"
            f.write(line)
    os.remove(xml_file)
```

```python

for file in os.listdir('labels'):
    filename = os.fsdecode(file)
    output = filename.split('.xml')[0] + '.txt'
    
    extract_bbox("labels/" + filename,"labels/" + output)
```
