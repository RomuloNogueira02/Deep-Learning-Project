```python
# Print some random images with or without labels
list_indexes = range(len(ds))
list_indexes = random.sample(list_indexes, 5)
print(list_indexes)
for i in list_indexes:
    img, label = ds[i]
    
    if len(label) > 0:
        print(str(i) + " have label")
        show_image_with_box(img, label)
    else:
        print(str(i) + " does not have label")
        show_image(img)


```
